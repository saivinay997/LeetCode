### 9. Palindrome Number 

Given an integer x, return true if x is a palindrome, and false otherwise.

```python
# My solution
def isPolindrome(x):
    if str(x) == str(x)[::-1]:
        return True
    else:
        return False
```

```python
# Leet peer solution. solution without turning the number into a string
def isPolindrome_1(x):
    if x<0:
        return False
    rev = 0
    orig = x
    # below code is for reversing any number
    while x != 0:
        rev = rev* 10 + x % 10
        x //= 10
    return rev == orig
```

```python
# leet solution. we don't need to reverse whole number

def isPolindrome_2(x):
    if x<0 or (x != 0 and x%10==0): 
        # here we are checking for negative numbers and tens(10,20,..) nums. 
        return False
    
    half = 0
    
    while x>half:
        half = half * 10 + x % 10 
        x = x//10
    return x == half or x == half//10 # addressing for both even and odd counts
```

### 13. Roman to Integer

Given a roman numeral, convert it to an integer.



```python
# my solution.
# it can also be done without converting to a list
def romanToInt(s):
    dict = {'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000}
    lst = []
    for i in s:
        lst.append(i)
    sum_roman = 0
    idx = 0
    while idx != len(lst):
        
        if idx == len(lst)-1 or (dict[lst[idx]] >= dict[lst[idx+1]]): 
            sum_roman += dict[lst[idx]]
            idx+=1
            
        elif dict[lst[idx]] < dict[lst[idx+1]]:
            sub_roman = abs(dict[lst[idx]] - dict[lst[idx+1]])
            sum_roman += sub_roman
            idx+=2
        
    return sum_roman
```

```python
# leet peer solution. 
def romanToInt_1(s):
    roman_to_integer = {'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000}
    s = s.replace("IV", "IIII").replace("IX", "VIIII").replace("XL", "XXXX").replace("XC", "LXXXX").replace("CD", "CCCC").replace("CM", "DCCCC")
    return sum(map(lambda x: roman_to_integer[x], s))
```

```python
# leet peer solution.
def romanToInt_2(s):
    dict = {'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000}
    sum_roman = 0
    for i in range(len(s)-1):
        if dict[s[i]] < dict[s[i+1]]:
            sum_roman -= dict[s[i]]
        else:
            sum_roman += dict[s[i]]
    return sum_roman + dict[s[-1]]
```

## 14. Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string " ".

```python
def longestCommonPrefix(strs):
    lst = []
    for i in strs[0]:
        for j in range(1,len(strs)-1):
            if i == strs[j][strs[0].index(i)]:
                lst.append(i)
            else:
                break
    return lst  
    
```
